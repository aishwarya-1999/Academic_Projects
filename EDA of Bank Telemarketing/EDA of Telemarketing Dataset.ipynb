## Data Cleaning

```python
df.head()
```

```python
# separating job and edu
df['job'] = df['jobedu'].apply(lambda x: x.split(',')[0])
df['education'] = df['jobedu'].apply(lambda x: x.split(',')[1])
df.head()
```

```python
df.drop('jobedu',axis=1, inplace=True)
df.head()
```

### Handling missing values

#### Age

```python
df.isnull().sum()
```

```python
100*(df.age.isnull().sum()/len(df))
```

```python
# Drop records with missing age
df = df[~df.age.isnull()]
```

```python
df.age.isnull().sum()
```

#### month

```python
df.month.isnull().sum()
```

```python
df.month.value_counts(normalize = True)
```

```python
month_mode = (df.month.mode()[0])
```

```python
df.month.fillna(month_mode, inplace=True)
```

```python
df.month.isnull().sum()
```

#### response

```python
df.response.isnull().sum()
```

```python
100*(df.response.isnull().sum()/len(df))
```

```python
df = df[~df.response.isnull()]
```

```python
df.response.isnull().sum()
```

#### pdays 
- According to the data description, '-1' indicates that the customer was never reached for any products.
- Changing '-1' to a missing value is important as it may impair the analysis.

```python
df.pdays.describe()
```

```python
df.loc[df.pdays<0,"pdays"] = np.NaN
df.pdays.describe()
```
